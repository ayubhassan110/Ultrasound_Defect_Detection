### Final Project Report (Executive Summary)

**Dataset:** 5000 simulated ultrasonic A‑scan signals, 5 defect classes (Normal, Crack, Porosity, Corrosion, Inclusion), each 500 points at 100 MHz sampling.

**Signal Processing Pipeline:** low‑pass, high‑pass, band‑pass Butterworth filters, Savitzky‑Golay smoothing, wavelet denoising (db4, soft thresholding), Min‑Max and Standard scaling.

**Feature Engineering:**
22 features extracted – 11 time‑domain (RMS, kurtosis, crest factor, etc.), 6 frequency‑domain (spectral centroid, entropy, bandwidth), 5 wavelet energy ratios.

**EDA:** PCA shows reasonable class separation; t‑SNE reveals distinct clusters for Crack and Porosity; top discriminative features are kurtosis, crest factor, and wavelet energy band 2.

**Machine Learning Results:** XGBoost achieved weighted F1 of 0.978 and ROC AUC 0.998 on a stratified hold‑out set. Feature importance highlights crest factor and spectral entropy as most predictive.

**Deep Learning Results:** 1D CNN achieved 98.2% test accuracy, outperforming LSTM (96.5%) and the hybrid (97.8%). The 2D CNN on spectrograms reached 96.0% with limited data.

**Key Findings:** Both hand‑crafted features and raw‑signal deep learning can accurately detect and classify defects. The CNN model can be deployed with minimal preprocessing.

**Future Improvements:** Incorporate transfer learning from real experimental data; investigate semi‑supervised learning for rare defect types; optimise for real‑time FPGA inference.
